```python
import seaborn as sns
import plotly.express as px

tips = sns.load_dataset("tips")

fig = px.histogram(tips, x="total_bill", nbins=30,
                   title="Distribution of Total Bill")

fig.show()
```

```python
import plotly.express as px
import seaborn as sns

tips = sns.load_dataset("tips")

fig = px.scatter(
    tips,
    x="total_bill",
    y="tip",
    color="sex",
    size="size",
    title="Total Bill vs Tip"
)

fig.show()
```

```python
avg_tip = tips.groupby("day")["tip"].mean().reset_index()

fig = px.bar(
    avg_tip,
    x="day",
    y="tip",
    title="Average Tip by Day",
    color="day"
)

fig.show()
```

```python
fig = px.box(
    tips,
    x="day",
    y="tip",
    color="day",
    title="Tip Distribution by Day"
)

fig.show()
```

```python
import plotly.express as px
import seaborn as sns

tips = sns.load_dataset("tips")

daily_bill = tips.groupby("day")["total_bill"].mean().reset_index()

fig = px.line(
    daily_bill,
    x="day",
    y="total_bill",
    markers=True,
    title="Average Bill by Day"
)

fig.show()
```

```python
fig = px.pie(
    tips,
    names="day",
    title="Customer Distribution by Day"
)

fig.show()
```

```python
import plotly.express as px

corr = tips.corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto=True,
    title="Correlation Heatmap"
)

fig.show()
```

```python
fig = px.scatter(
    tips,
    x="total_bill",
    y="tip",
    size="size",
    color="day",
    title="Bubble Chart: Bill vs Tip"
)

fig.show()
```

```python
fig = px.violin(
    tips,
    x="day",
    y="total_bill",
    color="day",
    box=True,
    points="all",
    title="Bill Distribution by Day"
)

fig.show()
```
